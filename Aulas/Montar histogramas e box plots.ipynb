```python
import pandas as pd
from matplotlib import pyplot as plt
from matplotlib import rcParams
import seaborn as sns

%matplotlib inline
rcParams['figure.figsize'] =5,4
sns.set_style('whitegrid')
```

```python
caminho = 'C:/Users/caiog/Downloads/Pythons/dados/mtcars.csv'
carros = pd.read_csv(caminho)
carros.columns = ['nomes','mpg','cyl','discp','hpt','drat','wt','qsec','vs','am','gear','carb']
mpg = carros['mpg']

plt.hist(mpg)
plt.show()
```

```python
sns.distplot(mpg)
```

```python
carros.plot(kind='scatter', x='hpt', y='mpg', c=['darkgray'], s=150)
```

```python
sns.regplot(x='hpt',y='mpg',data=carros,scatter=True)
```

```python
sns.pairplot(carros)
```

```python
colunas = ['mpg', 'discp','hpt','wt']
novo_df = pd.DataFrame((carros[colunas].values),columns=colunas)
carros_target = carros ['am'].values

novo_df['grupo'] = pd.Series(carros_target, dtype='category')
```

```python
sns.pairplot(novo_df,hue='grupo',palette='hls')
```

```python
carros.boxplot(column = 'mpg',by='am')
carros.boxplot(column = 'wt',by='am')
```

```python
sns.boxplot(x='am',y='mpg', data=carros, palette='hls')
```
