```python
import pandas as pd
import seaborn as sns
from matplotlib import pyplot as plt
from matplotlib import rcParams
from scipy.stats.stats import pearsonr

%matplotlib inline
rcParams['figure.figsize'] = 6,3
sns.set_style('whitegrid')

caminho = 'C:/Users/caiog/Downloads/Pythons/dados/mtcars.csv'

carros = pd.read_csv(caminho)
carros.columns = ['nomes','mpg','cyl','discp','hpt','drat','wt','qsec','vs','am','gear','carb']
```

```python
x = carros[['mpg','hpt','wt','qsec']]
sns.pairplot(x)
```

```python
mpg = carros['mpg']
hpt = carros['hpt']
wt = carros['wt']
qsec = carros['qsec']

coeficiente_pearson, p_valor = pearsonr(mpg,hpt)
print('coeficiente de pearson {0:3}'.format(coeficiente_pearson))
```

```python
coeficiente_pearson, p_valor = pearsonr(mpg,wt)
print('coeficiente de pearson {0:3}'.format(coeficiente_pearson))
```

```python
coeficiente_pearson, p_valor = pearsonr(mpg,qsec)
print('coeficiente de pearson {0:3}'.format(coeficiente_pearson))
```
