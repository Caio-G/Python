```python
import pandas as pd

from matplotlib import pyplot as plt
from matplotlib import rcParams
import seaborn as sns
```

### Criando gráficos de barras
#### Desenhando um gráfico de barras a partir de uma lista com matplotlib

```python
x = range(1, 10)
y = [1, 2, 3, 4, 0, 4, 3, 2, 1]

plt.bar(x, y)
```
