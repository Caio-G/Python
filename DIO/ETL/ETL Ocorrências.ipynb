```python
import pandas as pd
import pandera as pa
```

```python
df_full = pd.read_csv('ocorrencia_2010_2020.csv', sep=';', parse_dates=['ocorrencia_dia'])
df = df_full[['codigo_ocorrencia','codigo_ocorrencia2','ocorrencia_classificacao','ocorrencia_cidade','ocorrencia_uf','ocorrencia_aerodromo','ocorrencia_dia','ocorrencia_hora','total_recomendacoes']]
df
```

```python
df.dtypes
```

```python
schema = pa.DataFrameSchema({'codigo_ocorrencia' : pa.Column(pa.Int),
                           'codigo_ocorrencia2' : pa.Column(pa.Int),
                           'ocorrencia_classificacao' : pa.Column(pa.String),
                           'ocorrencia_cidade' : pa.Column(pa.String),
                           'ocorrencia_uf' : pa.Column(pa.String),
                           'ocorrencia_aerodromo' : pa.Column(pa.String),
                           'ocorrencia_dia' : pa.Column(pa.DateTime),
                           'ocorrencia_hora' : pa.Column(pa.String, nullable=True),
                           'total_recomendacoes' : pa.Column(pa.Int)})
schema.validate(df)
```
